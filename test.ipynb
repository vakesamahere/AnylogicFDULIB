```python
from collections import Counter
import math

# 示例数据列表
data_dict = normalized_dict

# 提取值和权重
values = [int(v) for v in data_dict.keys()]
weights = list(data_dict.values())

# 1. 算术平均值（Arithmetic Mean）
arithmetic_mean = sum(values) / len(values)

# 2. 几何平均值（Geometric Mean）
geometric_mean = math.prod(values) ** (1/len(values))

# 3. 调和平均值（Harmonic Mean）
# harmonic_mean = len(values) / sum(1/v for v in values)

# 4. 加权平均值（Weighted Mean）
weighted_mean = sum(data_dict[str(v)] * v for v in values) / sum(weights)

# 5. 众数（Mode）
mode = max(data_dict, key=data_dict.get)

# 6. 中位数（Median）
# 首先将字典转换为有序列表
sorted_items = sorted(data_dict.items(), key=lambda item: int(item[0]))
total_count = sum(weights)
median_index = total_count // 2

cumulative_count = 0
median = None
for value, count in sorted_items:
    cumulative_count += count
    if cumulative_count > median_index:
        median = value
        break
    elif cumulative_count == median_index and total_count % 2 == 0:
        next_value = next(sorted_items)[0]
        median = (int(value) + int(next_value)) / 2
        break

# 输出结果
print(f"算术平均值: {arithmetic_mean}")
print(f"几何平均值: {geometric_mean}")
# print(f"调和平均值: {harmonic_mean}")
print(f"加权平均值: {weighted_mean}")
print(f"众数: {mode}")
print(f"中位数: {median}")
```

```python
# 计算前百分之 x 的加权平均
def weighted_mean_top_percent(data_dict, percent):
    # 将字典按值排序
    sorted_items = sorted(data_dict.items(), key=lambda item: int(item[0]), reverse=False)
    
    # 计算前百分之 x 的数据点数量
    total_count = sum(data_dict.values())
    top_count = int(total_count * (percent / 100))
    
    cumulative_count = 0
    top_values = []
    top_weights = []
    
    for value, count in sorted_items:
        if cumulative_count >= top_count:
            break
        top_values.append(int(value)//7)
        top_weights.append(count)
        cumulative_count += count
    
    # 计算前百分之 x 的加权平均
    if sum(top_weights) == 0:
        top_weighted_mean = 0
    else:
        top_weighted_mean = sum(v * w for v, w in zip(top_values, top_weights)) / sum(top_weights)
    return top_weighted_mean, len(top_values)

# 计算并输出前百分之 x 的加权平均（x 从 50 到 100）
percent_values = []
weighted_means = []
sample_sizes = []

for percent in range(5, 101, 5):
    top_weighted_mean, sample_size = weighted_mean_top_percent(data_dict, percent)
    percent_values.append(percent)
    weighted_means.append(top_weighted_mean)
    sample_sizes.append(sample_size)
    print(f"前百分之 {percent} 的加权平均值: {top_weighted_mean}, 样本容量: {sample_size}")

# 绘制折线图
plt.figure(figsize=(10, 6))
plt.plot(percent_values, weighted_means, marker='o', label='Weighted average')
plt.plot(percent_values, sample_sizes, marker='x', label='Sample size')
plt.title('The weighted average and sample size of the top x percent')
plt.xlabel('Percentage (%)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# 初始化一个字典来存储合并后的结果
merged_data = {}

for value, weight in data_dict.items():
    # 计算 value 整除 7 后的结果
    key = (int(value) // 7)*7
    
    # 如果 key 已经存在于 merged_data 中，则合并 weight
    if key in merged_data:
        merged_data[key] += weight
    else:
        # 否则，将新的 key 和 weight 添加到 merged_data 中
        merged_data[key] = weight

# 将 merged_data 转换为两个列表：values 和 weights
values = list(merged_data.keys())
weights = list(merged_data.values())
for value, weight in zip(values, weights):
    print(f"{value}: {weight}")

plt.figure(figsize=(10, 6))
plt.bar(values, weights, width=3, align='center', label='Probability distribution')
plt.title('Probability distribution')
plt.xlabel('Value')
plt.ylabel('Weight')
plt.legend()
plt.grid(True)
plt.show()
```
